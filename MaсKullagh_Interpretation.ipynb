# Уравнения Эйлера

Рассмотрим задачу о движении твердого тела с неподвижной (закрепленной) точкой
О в инерциальной системе отсчета.
Тогда теорема об изменении кинетического момента запишется в
виде:

\begin{align}
\dot{\vec{K}_O} = \vec{M}_{O}^{ex}
\end{align}

Далее будем считать, что вектор угловой скорости тела
$\omega$
задается проекциями
$p, q, r$
на оси связанного с телом базиса
$O\vec{e_x} \vec{e_y} \vec{e_z}$,
т.е.

\begin{align}
\vec{\omega} = p\vec{e_x} + q\vec{e_y} + r\vec{e_z}
\end{align}

Тогда, учитывая, что тензор инерции
$J_O$
в
базисе
$O\vec{e_x} \vec{e_y} \vec{e_z}$
не меняется, производную
$\vec{K}_O$,
вычисляемую в
инерциальном базисе, можно представить суммой двух слагаемых:

\begin{align}
\dot{\vec{K}_O} = J_O\dot{\vec{\omega}} + \vec{\omega} \times J_O\vec{\omega}
\end{align}

где

\begin{align}
\dot{\vec{\omega}} = \dot{p}\vec{e_x} + \dot{q}\vec{e_y} + \dot{r}\vec{e_z}
\end{align}

Подставляя полученное уравнение в теорему об изменении кинетического момента получим диинамеические **Уравнения Эйлера** :

\begin{align}
\vec{M}_{O}^{ex} = J_O\dot{\vec{\omega}} + \vec{\omega} \times J_O\vec{\omega}
\end{align}

А в покоординатной записи, при выборе базиса, состоящего из главных осей твердого тела:

\begin{align}
A\dot{p} + (C - B)qr = M_1\\
B\dot{q} + (A - C)qr = M_2\\
C\dot{r} + (B - A)qr = M_3
\end{align}

Здесь $A, B, C$ - главные моменты инерции тела для неподвижной точки $O$,

$p,q,r$ – проекции угловой скорости тела на главные оси
инерции, а

$M_1, M_2, M_3$ – проекции момента внешних сил на главные
оси инерции.

Учитывая полную аналогию между
формулами, выражающими кинетические моменты относительно неподвижной точки тела и относительно его центра масс, и
дословно повторяя процедуру вывода уравнений, получим
следующие, аналогичные предыдущим, уравнения:
\begin{align}
\vec{M}_{O}^{ex} = J_O\dot{\vec{\omega}} + \vec{\omega} \times J_O\vec{\omega}
\end{align}
Здесь
$J_C$ – центральный тензор инерции тела, а
$\vec{M}_C^{ex}$
– момент действующих на тело внешних сил, вычисленный относительно центра
масс тела.



# Случай Эйлера

В задаче о движения твердого тела с неподвижной точкой $O$ **случай Эйлера** определяется условием
$
\vec{M_O} = 0$
и называется движением
тела по инерции.
В случае Эйлера сохраняется вектор кинематического момента в инерциальном базисе,
\begin{align}
\vec{K} = \vec{K}_O = \vec{const}
\end{align}
а также сохраняется кинетическая энергия тела $T$.

Исследуем движение **динамически симметричного тела в случае Эйлера**.

*Твердое тело называется динамически симметричным относительно точки $O$, если равны друг другу два из главных моментов инерции тела для точки $O$.*

Для динамически симметричного относительно точки $O$ тела тензор инерции в базисе главных осей инерции
$O\vec{e_x} \vec{e_y} \vec{e_z}$
имеет вид

\begin{align}
J_O =
\begin{pmatrix}
A & 0 & 0 \\
0 & A & 0 \\
0 & 0 & C
\end{pmatrix}
\end{align}

Угловая скорость твердого тела в проекциях на главные оси инерции записывается в виде
$\omega = p\vec{e_x} + q\vec{e_y} + r\vec{e_z}$
, а кинетический момент
для динамически симметричного тела выражается формулой
\begin{align}
\vec{K}_O = Ap\vec{e_x} + Bq\vec{e_y} + Cr\vec{e_z}
\end{align}

Выражая из этой формулы угловую скорость тела
\begin{equation}
    \vec{\omega} = \frac{\vec{K}_O}{A} + \frac{A - C}{A} r \vec{e}_z
\end{equation}

Угловая скорость раскладывается на две составляющие:
\begin{equation}
    \vec{\omega} = \vec{\omega}_1 + \vec{\omega}_2, \quad \vec{\omega}_1 = \frac{\vec{K}_O}{A}, \quad \vec{\omega}_2 = \frac{A - C}{A} r \vec{e}_z
\end{equation}

В рассматриваемой задаче составляющая $\vec{\omega}_1$ постоянна по величине и направлению. Составляющая $\vec{\omega}_2$ постоянна по величине, так как при \(A = B\) из третьего уравнения системы уравнений Эйлера следует $r = \text{const}$. Не меняется также и угол $\theta$ между векторами $\vec{\omega}_1$ и $\vec{\omega}_2$, поскольку имеет место равенство
\begin{align}
K_3 = Cr = K \cos \theta = \text{const}.
\end{align}

Таким образом движение симметричного тела в случае Эйлера является **регулярной прецессией**. Это движение представляется комбинацией двух простых вращений, где
первым является вращение вокруг неизменного направления кинетического момента с постоянной угловой скоростью прецессии
$\vec{\omega_1}$
,
вторым – вращение вокруг оси динамической симметрии тела с постоянной по величине угловой скоростью собственного вращения
$\vec{\omega_2}$
, а угол
$\theta$
между осями остается постоянным. Параметры этой регулярной прецессии определяются формулами:
\begin{equation}
    \dot{\psi} = \omega_1 = \frac{K}{A}, \quad \dot{\phi} = \omega_2 = \frac{A - C}{A} r, \quad \cos \theta = \frac{Cr}{K}
\end{equation}

Регулярная прецессия динамически симметричного тела в случае
Эйлера называется свободной, поскольку она происходит при отсутствии момента внешних сил. При этом, как следует из формул,
параметры свободной регулярной прецессии связаны равенством
\begin{align}
  C\omega_2 + (C - A)\omega_1cos\theta = 0
\end{align}
т. е. подбором начальных условий нельзя реализовать свободную регулярную прецессию с произвольными параметрами







*Визуализация прецессии твердого, динамически симметричного, тела в случае Эйлера*

```python
import numpy as np
from manim import *
config.media_embed = True

def generate_anim():
  render_code = f'''%%manim -pql -v WARNING MovingAngle

class MovingAngle(Scene):
    def construct(self):

      intro=Text("Движение динамически симметричного тела в случае Эйлера").scale(0.5)
      self.play(Create(intro))
      self.wait(2)
      self.play(FadeOut(intro))

      omega_1 = Vector([0, 3])
      w1 = MathTex(f"\\omega_1").next_to(omega_1.get_end(), RIGHT)
      self.play(Create(omega_1), Write(w1))
      self.play(omega_1.animate.set_color(BLUE))
      descrw1 = Text("Угловая скорость регулярной прецессии").scale(0.5).shift(DOWN)
      self.play(Write(descrw1))
      self.wait(2)

      omega_2 = Vector([2, 2])
      w2 = MathTex(f"\\omega_2").next_to(omega_2.get_end(), RIGHT)
      angle = Angle(omega_2, omega_1, radius=0.5, other_angle=False)
      angle_label = MathTex(r"\\theta").move_to(angle.get_center() + 0.3*UP + 0.1*RIGHT)
      self.play(FadeOut(descrw1), Create(omega_2), Create(angle), Write(angle_label), Write(w2))
      self.play(omega_2.animate.set_color(GREEN))
      descrw2 = Text("Угловая скорость собственного вращения").scale(0.5).shift(DOWN)
      self.play(Write(descrw2))
      self.wait(2)

      self.play(FadeOut(descrw2, angle, angle_label))
      omega_2_n = Vector([-2, 2], color=GREEN)
      self.play(Transform(omega_2, omega_2_n), w2.animate.move_to([-2.5,2,0]))
      new_angle = Angle(omega_1, omega_2, radius=0.5, other_angle=False)
      new_label = MathTex(r"\\theta = const").move_to(angle.get_center() + 1 * DOWN)
      self.play(Create(new_angle), Write(new_label))
      self.wait(2)
'''
  return render_code

from IPython import get_ipython
from IPython.display import display, clear_output

ipython = get_ipython()
render_code = generate_anim()
ipython.run_cell(render_code)
```

#  Интерпретация Мак-Кулага
В случае Эйлера уравнения вращательного движения твердого тела интегрируются при любых начальных условиях. Но получаемое при этом аналитическое решение выражается
довольно сложным образом через эллиптические функции и особой
наглядностью не отличается. В связи с этим анализ движения тела в
случае Эйлера дополняется геометрическими интерпретациями,
наиболее известными из которых являются интерпретации Пуансо и
Мак-Куллага.

В интерпретации Мак-Кулага первые интегралы записываются через проекции кинетического момента $\vec{K}$ на главные оси инерции тела в виде:
\begin{equation}
  K_1^2 + K_2^2 + K_3^2 = K^2\\
  \frac{K_1^2}{A} + \frac{K_2^2}{B} + \frac{K_3^2}{C} = 2T
\end{equation}
Здесь эллипсоид, описываемый вторым уравнением , называется эллипсоидом Мак-Куллага. Линии пересечения сферы, описываемой первым уравнением, с эллипсоидом представляют собой траектории неподвижного в инерциальном базисе вектора кинетического момента $\vec{K}$
на поверхности
неподвижного в теле эллипсоида Мак-Куллага. Эти траектории являются замкнутыми кривыми и разделены двумя сепаратрисами, пересекающимися по главной оси
$\vec{e}_2$
, соответствующей среднему по
величине из главных моментов инерции тела $(A < B < C)$.

Ellipsoid

# Эффект Джанибекова

Из рисунка можно видеть, что если в какой-то момент времени
вектор
$\vec{K}$
близок по направлению к оси среднего момента инерции
$\vec{e}_2$
, то в дальнейшем, двигаясь по соответствующей траектории, он
в какой-то момент времени займет на поверхности эллипсоида МакКуллага положение, близкое по направлению к оси $-\vec{e}_2$. Это означает, что по отношению к неподвижному в инерциальном базисе
вектору $\vec{K}$
ось $\vec{e}_2$
поменяет свое направление на обратное.
Если учесть еще, что скорость вектора
$\vec{K}$
относительно тела выражается формулой
$\dot{\vec{K}^{отн}} = \vec{K} \times \vec{J}^{-1}\vec{K}$
и она тем меньше, чем ближе
вектор
$\vec{K}$
по направлению к оси $\vec{e}_2$
или $- \vec{e}_2$
, то этого будет достаточно для объяснения так называемого «эффекта» Джанибекова, в
котором закрученное вокруг оси среднего момента инерции тело неоднократно и почти скачкообразно меняет направление этой оси в
пространстве на обратное.

*Здесь можно посмотреть на Эффект Джанибекова. Рассмотрено вращение главных осей тела в инерциальном базисе.*

```python
import numpy as np
from manim import *
config.media_embed = True

def generate_effect():
  render_code = f'''%%manim -pql -v WARNING DzhanibekovEffect

class DzhanibekovEffect(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=75 * DEGREES, theta=30 * DEGREES)
        basis_vectors = ThreeDAxes(
            x_length=3.5,
            y_length=3.5,
            z_length=3.5
        )
        basis_vectors.x_axis.set_color(BLUE)
        basis_vectors.y_axis.set_color(RED)
        basis_vectors.z_axis.set_color(GREEN)
        basis_vectors.add_updater(lambda mobject, dt: mobject.rotate(dt*720*DEGREES))
        self.add(basis_vectors)
        self.wait(2)
        self.play(
            Rotate(basis_vectors, angle=180 * DEGREES, axis=UP, about_point=ORIGIN),
            run_time=0.5,
        )
        self.wait(2)
        self.play(
            Rotate(basis_vectors, angle=180 * DEGREES, axis=UP, about_point=ORIGIN),
            run_time=0.5,
        )
'''
  return render_code

from IPython import get_ipython
from IPython.display import display, clear_output

ipython = get_ipython()
render_code = generate_effect()
ipython.run_cell(render_code)
```
